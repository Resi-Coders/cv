# GrayScale Transfortm

```python
from easycv import Image
from easycv.transforms.color import GrayScale
```

### Loading of the image

```python
img = Image("https://images.dog.ceo/breeds/germanshepherd/n02106662_5169.jpg")
```

### Showing the loaded image

```python
img
```

### Applying GrayScale Transform

```python
img = img.apply(GrayScale())
```

### Showing the image after the Transform

```python
img
```
