# Draw Examples

```python
from easycv import Image
from easycv.transforms.draw import Draw
```

For this example we will load an image with [random()](../../reference/image.rst#easycv.image.Image.random). For more examples of image loading check [Image examples](../image.ipynb).

```python
img = Image.random()
img
```

### Ellipse

To draw an ellipse apply the method Draw and select the method ellipse, for more information check [reference](../../reference/transforms/draw.rst#easycv.transforms.Draw.draw).

```python
img.apply(Draw(method="ellipse",
                ellipse=((200,100),100,60),
                rotation_angle=20,
                start_angle=0,
                end_angle=300,
                filled=False,
                color=(100,200,200),
                thickness=5,
                lineType="line_AA"))
```

A closed ellipse can also be drawn passing filled as True

```python
img.apply(Draw(method="ellipse",
                ellipse=((200,100),100,60),
                rotation_angle=20,
                start_angle=0,
                end_angle=300,
                filled=True,
                color=(100,200,200),
                thickness=5,
                lineType="line_AA"))
```

### Line

To draw a line apply the method Draw and select the method line, for more information check [reference](../../reference/transforms/draw.rst#easycv.transforms.draw.Draw).

```python
img.apply(Draw(method="line",
                pt1=(30,20),
                pt2=(200,150),
                color=(0,200,0),
                thickness=5,
                lineType="line_AA"))
```

### Polylines

To draw multiple lines or a polygon apply the method Draw and select the method polylines, for more information check [reference](../../reference/transforms/draw.rst#easycv.transforms.Draw.draw).

Not closed and not Filled

```python
img.apply(Draw(method = "polylines",
               points=[[50,50],[100,100],[50,100],[100,50]],
               color=(250,0,0),
               thickness=2,
               lineType="line_AA",
               isClosed=False,
               filled=False))
```

Closed and not Filled

```python
img.apply(Draw(method = "polylines",
               points=[[50,50],[100,100],[50,100],[100,50]],
               color=(0,0,250),
               thickness=2,
               lineType="line_AA",
               isClosed=True,
               filled=False))
```

Closed and Filled

```python
img.apply(Draw(method = "polylines",
               points=[[50,50],[100,100],[50,100],[100,50]],
               color=(100,0,70),
               thickness=2,
               lineType="line_AA",
               isClosed=False,
               filled=True))
```

### Rectangle

To draw a rectangle apply the method Draw and select the method rectangle, for more information check [reference](../../reference/transforms/draw.rst#easycv.transforms.Draw.draw).

Not Filled

```python
img.apply(Draw(method="rectangle", rectangle=((95,50),(350,250)),filled=True))
```

Filled

```python
img.apply(Draw(method="rectangle", rectangle=((95,50),(350,250)),color=(50,100,200),filled=False))
```

### Text

To draw Text apply the method Draw and select the method Text, for more information check [reference](../../reference/transforms/draw.rst#easycv.transforms.Draw.draw).

```python
img.apply(Draw(method = "text", text="easyCV", org=(1,200), font="SIMPLEX", size=4, color=(0,150,250),thickness=5))
```

The text can be mirrored using x_mirror

```python
img.apply(Draw(method = "text", text="easyCV", org=(1,200), font="SIMPLEX", size=4, color=(0,150,250),thickness=5,x_mirror=True))
```
