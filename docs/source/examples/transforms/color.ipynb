# Color Examples

```python
from easycv import Image
from easycv.transforms import GrayScale, Sepia, FilterChannels, GammaCorrection, Negative, Cartoon, PhotoSketch, ColorTransfer, ColorPick
```

For this example we will load an image with [random()](../../reference/image.rst#easycv.image.Image.random). For more examples of image loading check [Image examples](../image.ipynb).

```python
img = Image.random()
img
```

### Gray Scale

To convert the image to grayscale we just need to apply the transform, for more information check the [reference](../../reference/transforms/color.rst#easycv.transforms.color.GrayScale).


```python
img.apply(GrayScale())
```

### Sepia

To convert the image to grayscale we just need to apply the transform, for more information check the [reference](../../reference/transforms/color.rst#easycv.transforms.color.Sepia).

```python
img.apply(Sepia())
```

### Filter Channels

To remove a channel we just need to apply the transform with a list of the channels to remove, for more information check the [reference](../../reference/transforms/color.rst#easycv.transforms.color.FilterChannels).

```python
img.apply(FilterChannels(channels=[1]))
```

### Gamma Correction

To apply gamma correction we just need to apply the transform with a new gamma value, for more information check the [reference](../../reference/transforms/color.rst#easycv.transforms.color.GammaCorrection).

```python
img.apply(GammaCorrection(gamma=3.0))
```

```python
img.apply(GammaCorrection(gamma=0.5))
```

### Negative

To convert the image to negative we just need to apply the transform, for more information check the [reference](../../reference/transforms/color.rst#easycv.transforms.color.Negative).

```python
img.apply(Negative())
```

### Cartoon

To make an image look like a Cartoon we just need to apply the transform, for more information check the [reference](../../reference/transforms/color.rst#easycv.transforms.color.Cartoon).

```python
img.apply(Cartoon())
```

### Photo Sketch

To make an image look like a pencil sketch we just need to apply the transform, for more information check the [reference](../../reference/transforms/color.rst#easycv.transforms.color.PhotoSketch).

```python
img.apply(PhotoSketch())
```

### Color Transfer

To transfer the colors from source to target we just need to apply the transform, for more information check the [reference](../../reference/transforms/color.rst#easycv.transforms.color.ColorTransfer).

```python
src = Image("images/source.jpg")
src
```

```python
img.apply(ColorTransfer(source=src))
```

### Color Pick

To pick the color from the desired point or rectangle we just need to apply the transform, for more information check the [reference](../../reference/transforms/color.rst#easycv.transforms.color.ColorPick).

```python
img.apply(ColorPick(method="point"))
```
