# Filter Examples

```python
from easycv import Image
from easycv.transforms.filter import Blur, Sharpen, Sharpness
```

For this example we will load an image with [random()](../../reference/image.rst#easycv.image.Image.random). For more examples of image loading check [Image examples](../image.ipynb).

```python
img = Image.random()
img
```

### Blur

To blur an image we simply apply the Blur transform with the desired method, for more information check the [reference](../../reference/transforms/filter.rst#easycv.transforms.filter.Blur).

```python
img.apply(Blur(method="gaussian", size=11))
```

```python
img.apply(Blur(method="bilateral"))
```

### Sharpness

You can detect how sharpen an image image by applying the Sharpness transform, for more information check the [reference](../../reference/transforms/filter.rst#easycv.transforms.filter.Sharpness).

```python
img
```

```python
img.apply(Sharpness())
```

We can see that the image is sharpen. Now let's test on a blurry image.

```python
img_blurry = Image('images/blurred.jpg')
img_blurry
```

```python
img_blurry.apply(Sharpness())
```

### Sharpen

To Sharpen an image we simply apply the Sharpen transform, for more information check the [reference](../../reference/transforms/filter.rst#easycv.transforms.filter.Sharpen).

```python
img.apply(Sharpen())
```
