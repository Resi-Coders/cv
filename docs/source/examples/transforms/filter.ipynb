# Filter Examples

```python
from easycv import Image
from easycv.transforms.filter import Blur
```

For this example we will load an image from the [Dog API](https://dog.ceo/dog-api/). For more examples of image loading check [Image examples](../../reference/image.rst).

```python
img = Image("https://images.dog.ceo/breeds/pug/n02110958_13224.jpg")
img
```

### Blur

To blur an image we simply apply the Blur transform with the desired method, for more information check the [reference](../../reference/transforms/filter.rst#easycv.transforms.filter.Blur).

```python
img.apply(Blur(method="gaussian"))
```

```python
img.apply(Blur(method="bilateral"))
```
