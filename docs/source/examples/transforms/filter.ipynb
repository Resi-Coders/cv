# Filter Examples

```python
from easycv import Image
from easycv.transforms.filter import Blur, Sharpen
```

For this example we will load an image with [random()](../../reference/image.rst#easycv.image.Image.random). For more examples of image loading check [Image examples](../image.ipynb).

```python
img = Image.random()
img
```

### Blur

To blur an image we simply apply the Blur transform with the desired method, for more information check the [reference](../../reference/transforms/filter.rst#easycv.transforms.filter.Blur).

```python
img.apply(Blur(method="gaussian"))
```

```python
img.apply(Blur(method="bilateral"))
```

### Sharpen

To Sharpen an image we simply apply the Sharpen transform, for more information check the [reference](../../reference/transforms/filter.rst#easycv.transforms.filter.Sharpen).

```python
img = img.apply(Blur(method="gaussian"))
img.apply(Sharpen())
```
