# Edges Examples

```python
from easycv import Image
from easycv.transforms import Gradient, GradientAngle, GradientMagnitude, Canny
from easycv.io.output import show_grid
```

For this example we will load an image from the [Dog API](https://dog.ceo/dog-api/). For more examples of image loading check [Image examples].

```python
img = Image("https://images.dog.ceo/breeds/sheepdog-english/n02105641_1045.jpg")
img
```

### Gradient

To calculate the gradient of the image we simply apply the Gradient Transform with an axis, for more information check the [reference](../../reference/transforms/edges.rst)

```python
grad_x = img.apply(Gradient(axis="x"))
grad_y = img.apply(Gradient(axis="y"))
show_grid([grad_x,grad_y])
```

### Gradient Magnitude

To calculate the gradient magnitude of the image we simply apply the GradientMagnitude Transform with the size of the kernel, for more information check the [reference](../../reference/transforms/edges.rst)

```python
img.apply(GradientMagnitude(size=1))
```

### Gradient Angle

To calculate the gradient angles of the image we simply apply the GradientAngle Transform with the size of the kernel, for more information check the [reference](../../reference/transforms/edges.rst)

```python
img.apply(GradientAngle(size=1))
```

### Canny Edge

To get the edges of an Image we just apply the Canny Transform with the low and high threshold, for more information check the [reference](../../reference/transforms/edges.rst)

```python
img.apply(Canny(low=100, high=200))
```
