# Image examples

```python
from easycv.io import show_grid
from easycv import Image, Pipeline
from easycv.transforms import GrayScale, Gradient, Noise, Blur
```

### Loading an image

For this example we will load an image using [random()](../reference/image.rst#easycv.image.Image.random).

```python
img = Image.random()
img
```

### Image Propeties

We can check images properties such as width and height

```python
img.height, img.width
```

### Applying transforms/pipelines

We can apply any transforms to the image

```python
img_1 = img.apply(GrayScale())
img_2 = img.apply(Gradient())
show_grid([img_1, img_2])
```

For convenience, images above are displayed in a grid created using [show_grid()](../reference/io/output.rst).

We can also apply pipelines the same way

```python
img_3 = img.apply(Pipeline([Gradient(),Blur()]))
img_3
```

### Viewing Image

To view the image in a popup window call the show method

```python
img.show() # Image window pops up
```

### Lazy images

Lazy images delay computation while possible. To create a lazy image we just specify the lazy parameter

```python
img = Image.random()
img.loaded
```

Normally transforms/pipelines are executed in the moment. Lazy images will save them for later execution in a pipeline

```python
img = img.apply(Gradient())
img.pending
```

When any operation requires the updated image it will be loaded/computed. The changes are saved to prevent duplicated computation. You can force the computation with the compute method but calling any parameter/method (width, height, ...) that requires the updated image would have the same effect. Check the [reference](../reference/image.rst) for more information.

```python
img.compute()
```

```python
img.loaded
```

```python
img.pending
```

### Save

In the end we can save the image to a file

```python
img.save("Image.jpg")
```
